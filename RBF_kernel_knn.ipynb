### 100 1-dimensional points are considered, the points are typically from 1 to 100.

```python
# RBF kernel function
import numpy as np
import matplotlib.pyplot as plt
def rbf(a,b,sig):
    d = (a-b)**2
    return np.exp(-d/(2*(sig**2)))
```

```python
# Similarity matrix using RBF kernel for different sigma values.
f, ax = plt.subplots(2, 3, figsize=(16,8))
x = list(range(1,101))
for idx, sig in enumerate([0.1,1,5,10,20,40]):
    l = []
    for i in x:
        for j in x:
            l.append(rbf(i,j,sig))
    l = np.array(l)
    l = l.reshape(100,100)
    ax[idx//3, idx%3].imshow(l, cmap='hot', interpolation='nearest')
    ax[idx//3, idx%3].set_title(f'sigma: {sig}')
plt.show()
```

### Here we can see that as the value of sigma is increasing, more neighborhood points are contributing to the similarity value leading to a less sparse similarity matrix.
